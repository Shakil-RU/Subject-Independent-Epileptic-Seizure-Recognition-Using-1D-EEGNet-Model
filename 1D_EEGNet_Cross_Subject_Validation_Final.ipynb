```python
# ==============================================================================
# 1. IMPORTS
# ==============================================================================
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv1D, BatchNormalization, ELU, AveragePooling1D, Dropout
from tensorflow.keras.layers import DepthwiseConv1D, SeparableConv1D, Flatten, Dense
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from scipy.signal import butter, filtfilt
import matplotlib.pyplot as plt
import random
from sklearn.metrics import confusion_matrix, roc_curve, auc, classification_report
import seaborn as sns


# ==============================================================================
# 2. DATA PREPROCESSING & HELPER FUNCTIONS
# ==============================================================================
def apply_butter_filter(signal_data, lowcut, highcut, fs, btype):
    nyquist = 0.5 * fs
    low = lowcut / nyquist
    high = highcut / nyquist
    b, a = butter(4, [low, high], btype=btype)
    return signal_data.apply(lambda col: filtfilt(b, a, col), axis=0)

def data_cleaning(data):
    return apply_butter_filter(data, 0.5, 70, 250, 'band')

def powerline_removal(data):
    return apply_butter_filter(data, 49.0, 51.0, 250, 'bandstop')

def normalize(train_df, val_df, test_df):
    scaler = StandardScaler()
    scaler.fit(train_df)
    train_normalized = pd.DataFrame(scaler.transform(train_df), index=train_df.index, columns=train_df.columns)
    val_normalized = pd.DataFrame(scaler.transform(val_df), index=val_df.index, columns=val_df.columns)
    test_normalized = pd.DataFrame(scaler.transform(test_df), index=test_df.index, columns=test_df.columns)
    return train_normalized, val_normalized, test_normalized

def subset_blocks(signal_df, block_length, label_type, overlap=0.5, sample_rate=250):
    signal = signal_df.to_numpy()
    samples_per_block = int(block_length * sample_rate)
    overlap_samples = int(overlap * samples_per_block)
    step_size = samples_per_block - overlap_samples
    num_blocks = (signal.shape[0] - samples_per_block) // step_size + 1
    if num_blocks <= 0: return np.array([]), np.array([])
    subsetted_signal = np.empty([num_blocks, samples_per_block, signal.shape[1]])
    for block_idx in range(num_blocks):
        start = block_idx * step_size
        end = start + samples_per_block
        subsetted_signal[block_idx] = signal[start:end, :]
    labels = np.full(num_blocks, label_type)
    return subsetted_signal, labels

# ==============================================================================
# 3. TENSORFLOW MODEL DEFINITION: 1D EEGNet
# ==============================================================================
def build_strong_eegnet_1d(C=29, T=500, F1=32, D=2, F2=64, F3=128, dropout_rate=0.4):
    inputs = Input(shape=(T, C))  # (timesteps, channels)

    # Block 1: Temporal Convolution
    x = Conv1D(F1, kernel_size=64, padding='same', use_bias=False)(inputs)
    x = BatchNormalization()(x)

    # Block 2: Depthwise Spatial Convolution
    x = DepthwiseConv1D(kernel_size=C, depth_multiplier=D, padding='valid', use_bias=False)(x)
    x = BatchNormalization()(x)
    x = ELU()(x)
    x = AveragePooling1D(pool_size=4)(x)
    x = Dropout(dropout_rate)(x)

    # Block 3: Separable Convolution (Spatial-Temporal)
    x = SeparableConv1D(F2, kernel_size=16, padding='same', use_bias=False)(x)
    x = BatchNormalization()(x)
    x = ELU()(x)
    x = AveragePooling1D(pool_size=4)(x)
    x = Dropout(dropout_rate)(x)

    # Block 4: Deeper Separable Convolution
    x = SeparableConv1D(F3, kernel_size=8, padding='same', use_bias=False)(x)
    x = BatchNormalization()(x)
    x = ELU()(x)
    x = AveragePooling1D(pool_size=2)(x)
    x = Dropout(dropout_rate)(x)

    # Final Classification Layer
    x = Flatten()(x)
    x = Dense(64, activation='elu')(x)
    x = Dropout(dropout_rate)(x)
    outputs = Dense(1, activation='sigmoid')(x)

    model = Model(inputs=inputs, outputs=outputs)
    return model


# ==============================================================================
# 4. MAIN EXECUTION BLOCK
# ==============================================================================
if __name__ == '__main__':

    SEED = 42
    random.seed(SEED)
    np.random.seed(SEED)
    tf.random.set_seed(SEED)

    data = pd.read_csv('/content/drive/MyDrive/Dataset/seizure_nonseizure_with_subject_info.csv')
    unique_patients = data.iloc[:, 0].unique()
    print(f"Total patients: {len(unique_patients)}")

    all_fold_losses, all_fold_accuracies = [], []

    for i, test_patient in enumerate(unique_patients):
        print(f"\n{'='*25} Fold {i+1}/{len(unique_patients)} | Test Patient: {test_patient} {'='*25}")

        test_df = data[data.iloc[:, 0] == test_patient]
        train_df = data[data.iloc[:, 0] != test_patient]

        train_set = train_df.drop(columns=[train_df.columns[0]])
        test_set = test_df.drop(columns=[test_df.columns[0]])

        preictal_train_df = train_set[train_set['outcome'] == 0].iloc[:, :-1]
        ictal_train_df = train_set[train_set['outcome'] == 1].iloc[:, :-1]
        preictal_test_df = test_set[test_set['outcome'] == 0].iloc[:, :-1]
        ictal_test_df = test_set[test_set['outcome'] == 1].iloc[:, :-1]

        preictal_train_clean = powerline_removal(data_cleaning(preictal_train_df))
        ictal_train_clean = powerline_removal(data_cleaning(ictal_train_df))
        preictal_test_clean = powerline_removal(data_cleaning(preictal_test_df))
        ictal_test_clean = powerline_removal(data_cleaning(ictal_test_df))

        # This normalization fits on all training patients together
        preictal_train_norm, _, preictal_test_norm = normalize(preictal_train_clean, preictal_test_clean, preictal_test_clean)
        ictal_train_norm, _, ictal_test_norm = normalize(ictal_train_clean, ictal_test_clean, ictal_test_clean)

        preictal_train_win, preictal_train_labels = subset_blocks(preictal_train_norm, 2, 0)
        ictal_train_win, ictal_train_labels = subset_blocks(ictal_train_norm, 2, 1)
        X_train_full = np.concatenate((preictal_train_win, ictal_train_win), axis=0)
        y_train_full = np.concatenate((preictal_train_labels, ictal_train_labels), axis=0)

        preictal_test_win, preictal_test_labels = subset_blocks(preictal_test_norm, 2, 0)
        ictal_test_win, ictal_test_labels = subset_blocks(ictal_test_norm, 2, 1)
        X_test = np.concatenate((preictal_test_win, ictal_test_win), axis=0)
        y_test = np.concatenate((preictal_test_labels, ictal_test_labels), axis=0)

        # This split creates the flawed validation set from the training patients
        X_train, X_val, y_train, y_val = train_test_split(
            X_train_full, y_train_full, test_size=0.2, random_state=42, stratify=y_train_full
        )
        print(f"Train shape: {X_train.shape}, Val shape: {X_val.shape}, Test shape: {X_test.shape}")

        # --- MODEL CHANGED TO EEGNet1D ---
        model = build_strong_eegnet_1d()

        model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001), loss='binary_crossentropy', metrics=['accuracy'])

        callbacks = [
            EarlyStopping(monitor='val_loss', patience=40, restore_best_weights=True, verbose=1),
            ReduceLROnPlateau(monitor='val_loss', factor=0.3, patience=10, verbose=1, min_lr=1e-6)
        ]

        history = model.fit(
            X_train, y_train,
            validation_data=(X_val, y_val),
            epochs=150,
            batch_size=32,
            callbacks=callbacks,
            verbose=2
        )

        print(f"\n--- Evaluating Fold {i+1} on Test Patient {test_patient} ---")
        loss, acc = model.evaluate(X_test, y_test, verbose=0)

        print(f"Test Accuracy: {acc:.4f} | Test Loss: {loss:.4f}")
        all_fold_accuracies.append(acc)
        all_fold_losses.append(loss)

        # ==================================================================
        # ADDITIONAL EVALUATION METRICS & VISUALIZATIONS
        # ==================================================================

        # --- Predict on test set ---
        y_pred_proba = model.predict(X_test)
        y_pred = (y_pred_proba > 0.5).astype(int)

        # --- Confusion Matrix ---
        cm = confusion_matrix(y_test, y_pred)
        plt.figure(figsize=(5, 4))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
        plt.xlabel('Predicted Label')
        plt.ylabel('True Label')
        plt.title(f'Confusion Matrix - Fold {i+1} (Patient {test_patient})')
        plt.tight_layout()
        plt.show()

        # --- Classification Report (optional printout) ---
        print("\nClassification Report:")
        print(classification_report(y_test, y_pred, digits=4))

        # --- Accuracy & Loss Curves ---
        plt.figure(figsize=(10, 4))
        plt.subplot(1, 2, 1)
        plt.plot(history.history['accuracy'], label='Train Accuracy')
        plt.plot(history.history['val_accuracy'], label='Val Accuracy')
        plt.title(f'Accuracy Curve - Fold {i+1}')
        plt.xlabel('Epoch')
        plt.ylabel('Accuracy')
        plt.legend()
        plt.grid(True, linestyle='--', alpha=0.6)

        plt.subplot(1, 2, 2)
        plt.plot(history.history['loss'], label='Train Loss')
        plt.plot(history.history['val_loss'], label='Val Loss')
        plt.title(f'Loss Curve - Fold {i+1}')
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.legend()
        plt.grid(True, linestyle='--', alpha=0.6)
        plt.tight_layout()
        plt.show()

        # --- ROC & AUC Curve ---
        fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
        roc_auc = auc(fpr, tpr)

        plt.figure(figsize=(6, 5))
        # Plot ROC curve line
        plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC curve (AUC = {roc_auc:.4f})')

        # Fill area under the curve
        plt.fill_between(fpr, tpr, alpha=0.3, color='blue')

        # Random classifier line
        plt.plot([0, 1], [0, 1], 'k--', label='Random Classifier')

        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title(f'ROC Curve - Fold {i+1} (Patient {test_patient})')
        plt.legend(loc='lower right')
        plt.grid(True, linestyle='--', alpha=0.6)
        plt.tight_layout()
        plt.show()



    avg_acc = np.mean(all_fold_accuracies)
    std_acc = np.std(all_fold_accuracies)
    avg_loss = np.mean(all_fold_losses)

    print("\n\n" + "="*50)
    print("=== Overall Cross-Subject Performance ===")
    print(f"Average Test Accuracy: {avg_acc:.4f} (+/- {std_acc:.4f})")
    print(f"Average Test Loss: {avg_loss:.4f}")
    print("="*50)

    #ploting all fold accuracy (bar plot)
    patients = [f'Patient {i+1}' for i in range(len(all_fold_accuracies))]
    plt.figure(figsize=(12, 6))
    bars = plt.bar(patients, all_fold_accuracies, color='green')
    plt.xlabel('Patient')
    plt.ylabel('Accuracy')
    plt.ylim(0, 1)
    # plt.title('Cross Subject Validation Result')
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.6)

    # Annotate accuracy values on top of bars
    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2.0, yval + 0.01, f'{yval:.2f}', ha='center', va='bottom')

    plt.tight_layout()
    plt.show()
```
